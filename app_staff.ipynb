```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
from matplotlib import style
style.use('seaborn-darkgrid')
import matplotlib.image as mpimg
import numpy as np
```

```python
cbp = pd.read_csv("cbp_apprehensions.csv")
```

```python
cbp_df = pd.DataFrame(cbp)
cbp_df = cbp_df.astype({"FY YEAR": str})
```

```python
cbp_df.head()
```

```python
cbp_df.dtypes
```

```python
img = mpimg.imread("border.png",0)
```

```python
plt.figure(figsize=(20,10))
plt.plot(cbp_df["FY YEAR"], cbp_df["APPREHENSIONS"], linewidth=5, marker="h", markersize=12)
plt.xlabel('Fiscal Year')
plt.ylabel('Total Apprehensions')
plt.title('Total CBP Apprehensions at the Southwest Border by Fiscal Year')
#plt.rcParams.update({'font.size': 18})
#plt.savefig('apprehensions.png')
plt.show()
```

```python
monthly = pd.read_csv('monthly_apprehensions.csv')
```

```python
monthly_df = pd.DataFrame(monthly)

monthly_df.head()
```

```python
monthly_df.dtypes
```

```python
monthly_df['APPREHENSIONS'].max()
```

```python
plt.figure(figsize=(60,30))
plt.plot(monthly_df["DATE"], monthly_df["APPREHENSIONS"], linewidth=10, marker="h", markersize=12)
plt.xlabel('Year')
plt.ylabel('Total Apprehensions')
plt.title('Total Border Patrol Apprehensions at the Southwest Border by Month')
plt.rcParams.update({'font.size': 42})
plt.savefig('monthly_apprehensions.png')
#plt.xticks(rotation='vertical')
#plt.tight_layout()
unique_years, ind = np.unique(monthly_df["YEAR"].values,return_index=True)
plt.xticks(monthly_df.index[ind], unique_years, rotation=(45))
plt.tight_layout()
plt.text(186, 0.5, 'Border Fence Completed', horizontalalignment='center',verticalalignment='center', rotation="vertical")
plt.text(80, 0.5, 'Secure Fence Act Passed', horizontalalignment='center',verticalalignment='center', rotation="vertical")
plt.text(225, 0.5, 'Military Deployed', horizontalalignment='center',verticalalignment='center', rotation="vertical")
plt.show()
```

```python
staffing = pd.read_csv("bp_staffing.csv")
staffing_df = pd.DataFrame(staffing)
staffing_df.head()
```

```python
plt.figure(figsize=(20,10))
plt.plot(staffing_df["FY YEAR"], staffing_df["STAFFING"], linewidth=5, marker="h", markersize=12, color="darkolivegreen")
plt.xlabel('Fiscal Year')
plt.ylabel('Total Agents')
plt.title('Total Border Patrol Agents for Southwest Border Sectors')
plt.rcParams.update({'font.size': 6})
plt.savefig('staffing.png')
plt.show()
```

```python
average_apprehensions = monthly_df["APPREHENSIONS"].mean()

average_apprehensions
```

```python
staffing_df["STAFFING"].max()
```

```python
plt.figure(figsize=(30,15))
plt.bar(staffing_df["FY YEAR"], staffing_df["STAFFING"], color="darkolivegreen")
plt.xlabel('Fiscal Year')
plt.ylabel('Total Agents')
plt.title('Total Border Patrol Agents for Southwest Border Sectors')
plt.rcParams.update({'font.size': 20})
plt.savefig('staffing_bar.png')
plt.show()
```

```python
plt.figure(figsize=(60,30))
plt.plot(monthly_df["DATE"], monthly_df["APPREHENSIONS"], linewidth=10, marker="h", markersize=12)
plt.xlabel('Year')
plt.ylabel('Total de Aprehensiones ')
plt.title('Total de Aprehensiones de la Patrulla Fronteriza en la Frontera Furoeste Por Mes')
plt.rcParams.update({'font.size': 42})
#plt.savefig('monthly_apprehensions.png')
#plt.xticks(rotation='vertical')
#plt.tight_layout()
unique_years, ind = np.unique(monthly_df["YEAR"].values,return_index=True)
plt.xticks(monthly_df.index[ind], unique_years, rotation=(45))
plt.tight_layout()
plt.text(186, 0.5, 'Cerco fronterizo completado', horizontalalignment='center',verticalalignment='center', rotation="vertical")
plt.text(80, 0.5, 'Ley de cerco seguro aprobada', horizontalalignment='center',verticalalignment='center', rotation="vertical")
plt.text(225, 0.5, 'Militares desplegados', horizontalalignment='center',verticalalignment='center', rotation="vertical")
plt.show()
```
